```python
import pandas as pd
import numpy as np
import string
```

```python
stores_df = pd.read_csv('us_supermarkets.csv')
stores_df.columns = ['nothing', 'store_name', 'lat', 'lon']
stores_df = stores_df.drop(columns=['nothing'])
stores_df.head()
```

```python
sort_df = stores_df.sort_values(by='store_name', ascending=True)

sort_df.store_name = sort_df.store_name.apply(lambda x: x.lower())
sort_df.store_name = sort_df.store_name.apply(lambda x: x.translate(string.punctuation))


sort_df.store_name 
```
